*Introduction to Functions*:

    Functions are blocks of code that perform a specific task. They are used to break up complex code into smaller, more manageable parts, which can make your code easier to read and understand. Functions can also be reused multiple times throughout your code, which can save you a lot of time and effort.

    Functions are defined using the def keyword, followed by the name of the function and a set of parentheses. Inside the parentheses, you can list any arguments that the function needs in order to perform its task. These arguments are optional, but they can be useful if you need to pass data into the function from outside.

    Once you have defined a function, you can call it from anywhere in your code using its name and passing any necessary arguments. When the function is called, the code inside it is executed, and any values that it returns are passed back to the calling code.

```python
import random
```

```python
# Begin by creating a list of items that the adventurer can pick up along the way. 
# These items will be used to help the adventurer overcome obstacles and defeat ghosts. 
# Examples of items can be weapons, potions, keys, etc.
backpack = []
health_points = 10
```

```python
def encounter_ghost():
    global health_points
    """
    This function handles the encounter with a ghost. 
    The outcome of the battle is determined by a random number between 1 and 10.
    If the random number is less than or equal to 5, the adventurer defeats the ghost. In this case, print the message "You defeated the ghost!" 
    and return something that indicates the adventurer defeated the ghost.
    If the random number is greater than 5, the adventurer loses the battle. In this case, print "You lost the battle..."
    and return something that indicates the ghost defeated the adventurer.
    """
    print("You encounter a ghost!\nThe ghost attacks")
    random_number = random.randint(1, 10)

    #win or loose:
    if 1<= random_number <=5:
        print("You defeated the ghost!")
        return True
    else:
        print("You lost the battle!\nYou loose 2 health_points.")
        health_points -= 2
        print(f'Health points: {health_points}')
        if health_points <= 0:
            print("GAME OVER! You lost all you health points")
        return False

print("Checking ghost encounter logic:", encounter_ghost())
```

```python
# main function for the game
# Complete the function called "run_mansion()" that serves as the main function for the game. 
# Within "run_mansion()", prompt the user to choose a path to take at each intersection. 
# Each path should have its unique challenges and obstacles that the adventurer must overcome.
def run_mansion():
    global health_points
    print("Welcome to the Haunted Mansion!")
    
    """
    Simulates an adventure through a haunted mansion. The adventurer starts with 10 health points and no items.
    Prompt the user to choose between two paths: "left" or "right". 

    If they choose "left", a random event occurs. There is a 50% chance that the adventurer will find a potion and a 50% chance that they will 
    fall into a trap and lose 2 health points. If they find the potion, it is saved into the adventurer's items. 
    If they fall into a trap, 2 points are taken out of the adventurer's health points.

    If they choose "right", the "encounter_ghost()" function is called to handle the battle between the adventurer and the ghost. 
    If the adventurer wins, they find a key which is saved into the adventurer's items. If they lose, they lose 2 health points.
    Hint: remember to check what encounter_ghost() return to make know if they won or lost.

    If the adventurer chooses something other than "left" or "right", they are prompted to try again.

    If the adventurer's health points reach 0 or less, the message "Game over, you lost all your health points." is printed.

    If the adventurer has the key, they can unlock the door and find the Treasure. This message is printed "You unlocked the door and found the 
    Treasure! Congratulations!".
    If they don't have the key, they are prompted to find it from the beginning.

    """
    print("You enter the Haunted Mansion through a back window.\nYou find yourself in the main kitchen.")
    #create a loop to ask where to go
    #loop 1
    while True: 
        where_to_go = input("Are you going left or right?").strip().lower() 
        #give instructions what happened after turning left
        if where_to_go == 'left':
            #random 50\50 %
            if random.choice([True, False]):
                print("You turn left and enter the cooks quarters, you find a back of salt")
                backpack.append('Salt')
                print(f'Backpack: {backpack}')
            else: 
                print("You turn left and enter the cooks quarters.\nA black cat with glowing eyes jumps out at you hissing!\nYou suffer a minor heart attack and loose 2 health points") 
                health_points -= 2
                print(f'Health points: {health_points}')
                if health_points <= 0:
                    print("GAME OVER! You lost all you health points")
                    return
            break
        #give instructions on what happens after turning right
        elif where_to_go == 'right':
            if encounter_ghost() == True:
                print("You find a key!\nyou unlock the door and find the Treasure!")
                return
            else:
                break
        #give instructions to pick left or right
        else:
            print("please choose 'left' or 'right'!")
    #loop 2   
    while True:
        where_to_go_next = input("Where would you like to go next: left or right?")
        if where_to_go_next == 'left':
            if random.choice([True, False]):
                print("You turn left and enter a long dark hall way.\nYou find a dagger.")
                backpack.append('Dagger') 
                print(f'Backpack: {backpack}')
    
            else: 
                print("You turn left and enter a long dark hall way.\nThe suits of armor against the wall come to life and start fighting you.\nYou get stabbed in the arm and loose 2 health points") 
                health_points -= 2
                print(f'Health points: {health_points}') 
                if health_points <= 0:
                    print("GAME OVER! You lost all you health points")
                    return
            break
        #give instructions on what happens after turning right
        elif where_to_go_next == 'right':
            if encounter_ghost() == True:
                print("You find a key!\nyou unlock the door and find the Treasure!")
                return
            else:
                break
        #give instructions to pick left or right
        else:
            print("please choose 'left' or 'right'!")
#loop 3   
    while True:
        where_to_go_next = input("Where would you like to go next: left or right?")
        if where_to_go_next == 'left':
            if random.choice([True, False]):
                print("You turn left and enter an old dusty library.\nYou find a treasure map.")
                backpack.append('Treasure map') 
                print(f'Backpack: {backpack}')
    
            else: 
                print("You turn left and enter an old dusty library.\nYou try to turn on the lights but you get shocked! You loose 2 health points") 
                health_points -= 2
                print(f'Health points: {health_points}') 
                if health_points <= 0:
                    print("GAME OVER! You lost all you health points")
                    return
            break
        #give instructions on what happens after turning right
        elif where_to_go_next == 'right':
            if encounter_ghost() == True:
                print("You find a key!\nyou unlock the door and find the Treasure!")
                return
            else:
                break
        #give instructions to pick left or right
        else:
            print("please choose 'left' or 'right'!")
#loop 4   
    while True:
        where_to_go_next = input("Where would you like to go next: left or right?")
        if where_to_go_next == 'left':
            if random.choice([True, False]):
                print("You turn left and find a hidden door, you go through find a network of secret tunnels.\nYou find a torch")
                backpack.append('Torch') 
                print(f'Backpack: {backpack}')
    
            else: 
                print("You turn left and find a hidden door, you go through find a network of secret tunnels.\nYou find a torch, but when you try to grab it spike come out of the walls. You loose 2 health points") 
                health_points -= 2
                print(f'Health points: {health_points}') 
                if health_points <= 0:
                    print("GAME OVER! You lost all you health points")
                    return
            break
        #give instructions on what happens after turning right
        elif where_to_go_next == 'right':
            if encounter_ghost() == True:
                print("You find a key!\nyou unlock the door and find the Treasure!")
                return
            else:
                break
        #give instructions to pick left or right
        else:
            print("please choose 'left' or 'right'!")
#loop 5   
    while True:
        where_to_go_next = input("Where would you like to go next: left or right?")
        if where_to_go_next == 'left':
            if random.choice([True, False]):
                print("You turn left and step out into a circular room with a stone altar in the center.\nYou walk to the altar and find gemstones strewn on top of it.")
                backpack.append('Gemstones') 
                print(f'Backpack: {backpack}')
    
            else: 
                print("You turn left and step out into a circular room with a stone altar in the center.\nYou walk to the altar and step on a trap stone.\nWith a bang a huge bould closes the entrence you just came through. You loose 2 health point") 
                health_points -= 2
                print(f'Health points: {health_points}') 
                if health_points <= 0:
                    print("GAME OVER! You lost all you health points")
                    return
          #give instructions on what happens after turning right
        elif where_to_go_next == 'right':
            if encounter_ghost() == True:
                print("You find a key!\nYou unlock the door and find the Treasure!")
                return
            else:
                break
        #give instructions to pick left or right
        else:
            print("please choose 'left' or 'right'!")

```
